## Группировка и агрегация

`.groupby()` позволяет осуществлять группировку в рамках столбца (или нескольких столбцов)

```python
# Посчитать сколько пассажиров было какого возраста 
# Группируем по полю Age
# PassengerId - уникальное значение для каждой строки, следовательно будем считать их
frame.groupby('Age')["PassengerId"].count()
```

```python
# Можно убедиться, что строк с возврастом 0 ровно столько, сколько получилось при группировке
frame[frame.Age == 0.0]
```

```python
# Можно использовать метод у результата группировки agg(), перечислив функции, которые хотим определить

# count - кол-во объектов
# min - минимальное значение в группе
# max - максимальное значение в группе

frame.groupby("Pclass").Age.agg(['count', 'max', 'min'])
```

```python
# Когда группируем сразу по нескольким столбцам, получается Multi-Index
# С которым сложно будет работать 
group_res_one = frame.groupby(["Pclass", "isAlone"]).Fare.agg(['count', 'max', 'min'])
print(group_res_one.index)
group_res_one
```

```python
# Позволяет сделать reset для индексов
group_res_one = group_res_one.reset_index()
print(group_res_one.index)
group_res_one
```

```python
# Можно выполнять агрегацию сразу по нескольким столбцам

# Пример: в рамках каждого класса, определить средний возраст, максимальную цену билета
frame.groupby('Pclass').agg({
    "Age":'mean',
    "Fare":"max"
})
```

```python
# Если хотим определить еще какое-то значение, то указываем как массив строк

# Пример: В рамках класса определить средний возраст, максимальный возраст в каждой группе
# максимальную цену билета
frame.groupby('Pclass').agg({
    "Age":['mean', 'max'],
    "Fare":"max"
})
```

```python
# Тоже самое можно сделат если происходит агрегация по нескольким столбцам
frame.groupby(['Pclass', 'isAlone']).agg({
    "Age":['mean', 'max'],
    "Fare":"max"
})
```

```python
# Можно задать свои имена для колонок, для удобства
new_agg_frame = frame.groupby(['Pclass', 'isAlone']).agg(
    mean_age=('Age', 'mean'),
    max_age=('Age', 'max'),
    max_fare=("Fare", "max")
)
new_agg_frame
```

## Соединение нескольких фреймов

```python
# Предположим, что нужно соединить 2 DataFrame

df_one = pd.DataFrame({
    'id':[0, 1, 2, 3, 4],
    'prod':['A', "B", "C", "D", "E"],
    'price':[101, 102.3, 32, 403.2, 345.5]
})
print(df_one)

df_two = pd.DataFrame({
    'id': [4, 2, 3, 1, 0, 5],
    "last_price": [405.5, 52, 370.2, 132.3, 111, 99]
})
print("=====")
print(df_two)
```

```python
# Можно соединить по общему столбцу (Похоже на соединение таблиц в SQL)
# on='id' - по какому полю
# how='inner' - как соединять
df_3_merge = pd.merge(df_one, df_two, on='id', how='inner')

# (можно заметить, что столбцы объединились по полю 'id', к фрейму df_one, был добавлен df_two по полю 'id')
df_3_merge

# Вместо how='inner' можно использовать how='left', how='right', результат будет изменять
```

```python
# Аналог
df_one.merge(df_two, on='id', how='right')
```

```python
# Можно сделать конкатенацию фреймов
# axis=0 - добавляет строчки
# axis=1 - добавляет столбцы
# ignore_index=True - игнорировать ли исходные индексы в датафреймах

pd.concat([df_one, df_two], axis=0, ignore_index=True)
```

```python
# Можно заметить разный порядок столбцов
pd.concat([df_one, df_two], axis=1, ignore_index=False)
```
